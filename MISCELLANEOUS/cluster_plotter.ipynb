```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('../OUTPUTS/random_points_clusters.csv')  # replace 'path_to_your_file.csv' with the actual file path

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Get unique clusters
clusters = sorted(df['cluster'].unique())

# Color map
colors = plt.cm.jet(np.linspace(0, 1, len(clusters) - 1)) if len(clusters) > 1 else []

for i, cluster in enumerate(clusters):
    # Filter points belonging to the current cluster
    points = df[df['cluster'] == cluster]
    
    if cluster == -1:
        color = 'black'
    else:
        color_index = i-1 if -1 in clusters else i
        color = colors[color_index]
    
    # Plot points with the cluster-specific color
    ax.scatter(points['x'], points['y'], points['z'], s=20, color=color, label=f'Cluster {cluster}')

# Labeling
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title('3D Cluster Plot')
plt.show()
```
