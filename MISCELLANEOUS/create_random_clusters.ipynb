```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
NUM_CLUSTERS = 20

list_of_points = []
for c in range(NUM_CLUSTERS):
    x_cluster = np.random.uniform(-50,50)
    y_cluster = np.random.uniform(-50,50)

    for i in range(1000):
        list_of_points.append(np.random.multivariate_normal([x_cluster,y_cluster,0], [[1,0,0],[0,1,0],[0,0,3]]))

points = np.array(list_of_points)
points_x = points[:,0]
points_y = points[:,1]
points_z = points[:,2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(points_x, points_y, points_z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
from sklearn.cluster import DBSCAN
from matplotlib import colormaps

clustering = DBSCAN(eps=2, min_samples=3).fit(points)
clustering.labels_

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.scatter(points_x, points_y, points_z, c=clustering.labels_)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import pandas as pd

df = pd.DataFrame()
df["x"] = points_x
df["y"] = points_y
df["z"] = points_z
df["cluster"] = clustering.labels_

df_index = list(range(len(df)))
np.random.shuffle(df_index)
df = df.iloc[df_index]
df

df.to_csv("random_points_clusters.csv", index=False)

fd = pd.DataFrame()
fd["x"] = points_x
fd["y"] = points_y
fd["z"] = points_z

fd_index = list(range(len(fd)))
np.random.shuffle(fd_index)
fd = fd.iloc[fd_index]
fd

fd.to_csv("random_points.csv", index=False)
```
